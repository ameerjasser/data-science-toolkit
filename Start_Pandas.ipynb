# Accessing Data with Pandas

```python
coffee # or print(coffee)   or display(coffee)
```

```python
coffee.sample(10)  # Ten random rows
```

```python
# Accessing specific data

coffee.loc[0]   # coffee.loc[ #Rows, #Columns]
```

```python
coffee.loc[[0,1,2]]  # Access the first and second and third location (first three rows)
```

```python
coffee.loc[0:3]  # Notic that 0 is inclusive as well as 3 is inclusive

coffee.loc[0:]  # from 0 till end

coffee.loc[5:]  # from 5 till end

coffee.loc[5:9]  # from 5 till 9


# As we said before in this example just will appear the output of last line of code
```

```python
coffee.loc[5:9, ["Day", "Units Sold"]]  # This will appear just the columns( Day and Units sold) in the rows from 5 till 9 .
```

```python
coffee.iloc[0:2, 0:2] # Notice here we used iloc to input index istead of the name of columns
                      # And Notice that 2 is exclusive in this line of code
                      # (iloc) It's only use index values
```

```python
coffee.index = coffee["Day"]  # We can access the rows now deponds in the day
```

```python
coffee.head()
```

```python
coffee.loc["Monday"]  # This will appear the rows that in Day(Monday)
```

```python
coffee.loc["Monday":"Wednesday", "Units Sold"]
```
