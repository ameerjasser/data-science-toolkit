```python
# Step 1: Create a simple dataframe with obvious outliers
data = {
    'Score': [85, 90, 78, 92, 88, 200, 85, 89, -50, 87]
}
df = pd.DataFrame(data)

print("ORIGINAL DATA: ")
print(df)

# Step 2: Romove outliers using Z-score method
df_clean = df[(np.abs(stats.zscore(df)) < 2).all(axis=1)]

print("\nCLEANED DATA (outliers removed): ")
print(df_clean)

# Step 3: Show what happened
print(f"\nSUMMARY: ")
print(f"Original rows: {len(df)}")
print(f"Rows after cleaning: {len(df_clean)}")
print(f"Outliers removed: {len(df) - len(df_clean)}")
```
