```python
#Convert "Salary" string to integers
df['Salary'] = df['Salary'].replace(',','',regex=True)
#This replaces the comma ',' in 55,000 to 55000

df['Salary'] = df['Salary'].astype(int)
#This converts the column datatypes from object to numbers, in this case integer

df.head()
#This displays the first  rows to verify the conversion
```

```python
# Use df.info to identify which columns has a lot of missing values
df.info()
```

```python
missing_compensation = df['Compensation'].isna().sum()
print(f"Number of rows with missing calues in 'Compensation' column: {missing_compensation}")
```

```python
def missing_values_summary(df):
  """
  Returns each column name with the number of missing values

  Parameters:
  df (pd.DataFrame): Input dataframe

  Returns:
  pd.Series: Column names with missing value counts
  """

  missing_counts = df.isna().sum()
  return missing_counts[missing_counts >0] #Return only columns with missing values

# Usage
missing_values = missing_values_summary(df)
print("Columns with missing values: ")
print(missing_values)
```

```python
#Drop colums 'Jop title context', 'Other currency', 'Salary context'
df = df.drop(['Job title context', 'Other currency', 'Salary context'], axis=1)

#Check using df.head()
df.head()
```

```python
# Number of rows before
rows_before = len(df)
print(f"Rows before: {rows_before}")

# Drop rows with any missing values
df = df.dropna()

# Number of rows after
rows_after = len(df)
print(f"Rows after: {rows_after}")
print(f"Rows dropped: {rows_before - rows_after}")

# Check the result
df.head()
```

```python
# Number of rows before
rows_before = len(df)
print(f"Rows before: {rows_before}")

# Drop_rows with more than  missing values
df = df.dropna(thresh=2)

# Number of rows after
rows_after = len(df)
print(f"Number of rows after: {rows_after}")
print(f"Rows deleted: {rows_before - rows_after}")

df.head()
```

```python
# Check for each columns. Example is 'Gender'
# Write a code that prints out each columns and their number of  unique items

print('--Columns and their number of unique items--')
for column in df.columns:
  n_unique = df[column].nunique()
  print(f'{column} : {n_unique}')
```
